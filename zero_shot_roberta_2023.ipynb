```python
from transformers import pipeline
classifier = pipeline('zero-shot-classification', model='roberta-large-mnli')
```

```python
sequence_to_classify = ["The Mexican President has Tested Positive for COVID, Despite Saying Magical Religious Stamps Would Protect him.","Sex is no cure for Covid-19, says Health Department"]
candidate_labels = ['humor', 'not humor']
classifier(sequence_to_classify, candidate_labels)
```

```python
from google.colab import files
 
uploaded = files.upload()
```

```python
import pandas as pd
import io
 
df = pd.read_csv(io.BytesIO(uploaded['test-set.csv']),encoding="ISO-8859-1" )
print(df)
```

```python
sequence_to_classify = df["text"]
candidate_labels = ['humor', 'not humor']
scores = []
for i in sequence_to_classify:
  output = classifier(i, candidate_labels)
  # print(output)
  if output['labels'][0] == 'not humor':
    scores.append(0)
  elif output['labels'][0] == 'humor':
    scores.append(1)
```

```python
print(scores)
print(len(scores))
```

```python
import csv
with open("zero-shot-roberta.csv", 'w', newline='') as myfile:
     wr = csv.writer(myfile)
     for i in scores:
      wr.writerow([i])
```
